The <a href="http://en.wikipedia.org/wiki/IceCube_Neutrino_Observatory">Icecube</a> experiment is designed to detect cosmic ray <a href="http://en.wikipedia.org/wiki/Neutrino">neutrinos</a>. It is located in the Antarctic.

```python
from IPython.display import Image
Image(filename='IceCube_Neutrino_Observatory.jpg',width=400)
# Image credit: NSF/S. Lidstrom
```

Some neutrinos will hit atoms in the ice and produce other particles which light up the deep, otherwise dark ice. Icecube has melted holes in the ice and lowered strings of detectors to record this light and thereby reconstruct the neutrino which produced this collision. Icecube has instrumented roughly a cubic kilometer of the ice!

```python
Image(filename='IceCube_standalone.jpg',width=500)
# Credit: ICE.WUSTL.EDU/Pete Guest.
```

Icecube has released a subset of their dataset for others to analyze and play with. It contains information about which detectors recorded the light, how much light was recorded, and where these detectors are located. Let's take a look at it!

First we need to import our helper functions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Uncomment the following line if you want your plots to be displayed in a separate interactive window.
#%pylab qt 

from icecube_tools import get_collisions,display_icecube_detectors,get_detector_info
```

Let's grab the neutrino collisions in the file, see how many we have, and take a look at a few.

```python
infile = open('small_icecube_sample.odf')

# Get the neutrino collisions!
collisions = get_collisions(infile)

print "# of collisions: %d" % (len(collisions))
display_icecube_detectors(collisions[0])
```

```python
display_icecube_detectors(collisions[1])
```

```python
display_icecube_detectors(collisions[2])
```

Or try analyzing the data yourself!

```python
energies = []
for collision in collisions:
    
    q,t,x,y,z = get_detector_info(collision)
    
    energies.append(sum(q))
    
h = plt.hist(energies)
```
